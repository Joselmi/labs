## Sección 7 - Estructuras complejas y Agregación

La siguiente celda es de solo lectura e insertará una lista de clientes que será requerida para las preguntas posteriores

````python
import random
import faker
import faker_commerce
from mimesis import Generic
from datetime import datetime, timedelta, timezone
from typing import cast
from pprint import pprint

fake = faker.Faker("es_MX")
fake.add_provider(faker_commerce.Provider)
mim_generic = Generic(locale="es_MX")

#######################

productos = []
for _ in range(random.randint(50, 100)):
    producto = {
        "producto": cast(faker_commerce.Provider, fake.unique).ecommerce_name(),
        "categoria": cast(faker_commerce.Provider, fake).ecommerce_category(),
        "precio": cast(faker_commerce.Provider, fake).ecommerce_price(as_int=False)
        / 5000.0,
    }
    productos.append(producto)

database_name = "laboratorio"
collection_name = "productos"
mongo_client[database_name].drop_collection(collection_name)
mongo_client[database_name][collection_name].insert_many(productos)
print(f'{len(productos)} documentos insertados en colección "{collection_name}"')

#######################

clientes = []
for _ in range(random.randint(500, 1000)):
    client = {
        "nombre": cast(faker.Faker, fake.unique).name(),
        "edad": random.randint(18, 65),
        "contacto": {
            "telefono": cast(faker.Faker, fake.unique).phone_number(),
            "email": cast(faker.Faker, fake.unique).email(),
        },
        "compras": [],
    }
    for _ in range(random.randint(1, 5)):
        producto = random.choice(productos)
        cantidad = random.randint(1, 3)
        compra = {
            "producto_id": producto["_id"],
            "fecha": fake.past_datetime(
                start_date=datetime.now(tz=timezone.utc) - timedelta(days=365)
            ),
            "cantidad": cantidad,
            "monto": producto["precio"] * cantidad,
        }
        client["compras"].append(compra)
    clientes.append(client)

database_name = "laboratorio"
collection_name = "clientes"
mongo_client[database_name].drop_collection(collection_name)
mongo_client[database_name][collection_name].insert_many(clientes)
print(f'{len(clientes)} documentos insertados en colección "{collection_name}"')
````
